```python
data_math = pd.read_csv('2017_2018_MATH.csv')
```

```python
data_math
```

```python
data_math_all_students = data_math[(data_math['Category'] == 'All Students')]
data_math_all_students = data_math_all_students[data_math_all_students['Mean Scale Score']!= 's']
```

```python
# Make all columns float or int as is necessary
data_math_all_students['Mean Scale Score'] = data_math_all_students['Mean Scale Score'].astype(int)
data_math_all_students['# Level 1'] = data_math_all_students['# Level 1'].astype(int)
data_math_all_students['# Level 2'] = data_math_all_students['# Level 2'].astype(int)
data_math_all_students['# Level 3'] = data_math_all_students['# Level 3'].astype(int)
data_math_all_students['# Level 4'] = data_math_all_students['# Level 4'].astype(int)

data_math_all_students = data_math_all_students[data_math_all_students['Grade'] != 'All Grades']
```

```python
data_math_all_students['Grade']  = data_math_all_students['Grade'].astype(int)
data_math_all_students['3to5'] = data_math_all_students['Grade'].apply(lambda x: (x < 6) & (x > 2) )
data_math_all_students['6to8'] = data_math_all_students['Grade'].apply(lambda x: (x < 9) & (x > 5) )
```

```python
data_math_all_students = data_math_all_students.groupby([data_math_all_students['DBN'], data_math_all_students['3to5'], data_math_all_students['6to8']]).sum()
```

```python
data_math_all_students['Sum'] = data_math_all_students['# Level 1'] + data_math_all_students['# Level 2'] + data_math_all_students['# Level 3'] + data_math_all_students['# Level 4']
data_math_all_students['% Level 1'] = 100*data_math_all_students['# Level 1']/data_math_all_students['Sum']
data_math_all_students['% Level 2'] = 100*data_math_all_students['# Level 2']/data_math_all_students['Sum']
data_math_all_students['% Level 3'] = 100*data_math_all_students['# Level 3']/data_math_all_students['Sum']
data_math_all_students['% Level 4'] = 100*data_math_all_students['# Level 4']/data_math_all_students['Sum']

# data_math_all_students.reset_index(inplace = True)
```

```python
# data_math_all_students
data_filtered = data_math_all_students.drop(columns=['3to5', '6to8']).reset_index()[['DBN', '3to5', '6to8', '% Level 1', '% Level 2', '% Level 3', '% Level 4']]

data_filtered_3to5 = data_filtered[data_filtered['3to5'] == True]
data_filtered_6to8 = data_filtered[data_filtered['6to8'] == True]
```

```python
data_filtered_3to5.drop(columns=['3to5', '6to8']).to_csv('Math_scores_3to5.csv')
data_filtered_6to8.drop(columns=['3to5', '6to8']).to_csv('Math_scores_6to8.csv')
```
